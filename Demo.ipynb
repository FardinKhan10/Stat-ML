```python
# what is the upper and lower fence values for tatal bill column

# iqr
q1 = data['tip'].quantile(0.25)
q3 = data['tip'].quantile(0.75)
iqr = q3 - q1

# fences
lower_fence = q1 - 1.5*iqr
upper_fence = q3 + 1.5*iqr
print('Lower Fence:', lower_fence)
print('Upper Fence:', upper_fence)
```

```python
# Q18: Count the number of palindromic substrings
# Input: 'aaa'
# Output: 6

s = "aaa"
count = 0

for i in range(len(s)):
  for j in range(i,len(s)):
    strr = s[i:j+1]
    if strr == strr[::-1]:
      count+=1

print(count)
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

m = len(matrix)       # number of rows
n = len(matrix[0])    # number of columns

top = 0
bottom = m - 1
left = 0
right = n - 1

result = []

while top <= bottom and left <= right:
    # Traverse from Left to Right
    for i in range(left, right + 1):
        result.append(matrix[top][i])
    top += 1

    # Traverse from Top to Bottom
    for i in range(top, bottom + 1):
        result.append(matrix[i][right])
    right -= 1

    # Traverse from Right to Left (only if rows left)
    if top <= bottom:
        for i in range(right, left - 1, -1):
            result.append(matrix[bottom][i])
        bottom -= 1

    # Traverse from Bottom to Top (only if columns left)
    if left <= right:
        for i in range(bottom, top - 1, -1):
            result.append(matrix[i][left])
        left += 1

# Print result
for num in result:
    print(num, end=" ")
```
